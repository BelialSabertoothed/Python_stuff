```python
import matplotlib.pyplot as plt
import numpy as np

def make_graph(temperatures):
    cities = list(temperatures.keys())
    days = range(1, 8)
    
    fig, ax = plt.subplots(figsize=(5, 4))
    
    for i, city in enumerate(cities):
        ax.plot(days, temperatures[city], label=city)
    
    ax.set_ylim(5, 30)
    ax.set_xlabel("Days")
    ax.set_ylabel("Temperature (Â°C)")
    ax.legend(loc='upper right')
    ax.set_xticks(days)
    ax.set_xticklabels(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
    plt.set_cmap('tab10')
    
    return fig
```

```python
temperatures = {
    "Prague": [12, 14, 15, 14, 13, 12, 11],
    "Tokyo": [18, 20, 19, 21, 22, 20, 18],
    "New York": [7, 8, 12, 10, 11, 9, 8],
    "Sydney": [22, 23, 25, 24, 23, 22, 21],
}

fig = make_graph(temperatures)

plt.show()
```
